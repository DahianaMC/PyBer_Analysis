```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Create line charts
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# MATLAB like interface
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Object-Oriented Interface
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot (MATLAB)
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Change the size of the figure in pixels.
plt.figure(figsize=(20, 10))
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Object-Oriented Interface
# Change the line color to green.
# Change the marker to a diamond.
# Change the line width to 2 points.
ax = plt.axes()
ax.plot(x_axis, y_axis, marker='D', color='green', linewidth=2, label='Boston')
# Add a legend for the city of Boston.
ax.legend()
# Add a title and axes labels.
ax.set_xlabel('Date')
ax.set_ylabel('Fare $()')
ax.set_title("PyBer Fare by Month")
# Add a y-axis limit.
ax.set_ylim(0, 45)
# Add grid lines.
ax.grid()
```

```python
# Create the plot.
# Switch axis, Fare($) data is on the x axis.
plt.scatter(y_axis, x_axis, color="red", label='Chicago')
# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Invert y axis
plt.gca().invert_yaxis()
# Add a limit for the x axis data.
plt.xlim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create a Bubble Chart Using the MATLAB Method
# s=size of the marker
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Making the marker bigger
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

# Use the new array to make the marker bigger
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Using list comprehension to replace many for and while loops
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create the plot using the Object-Oriented Interface.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c='skyblue', edgecolors='black', label='Boston', alpha=0.8, s=[i*5 for i in y_axis])
ax.invert_yaxis()
# Create labels for the x and y axes.
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
# Create a title.
ax.set_title('PyBer Fare by Month')
# Add the legend.
ax.legend()
# Add a limit
ax.set_xlim(0,50)
```

```python
# Create a Pie Chart Using the MATLAB Method
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Increase pie size
plt.subplots(figsize=(8, 8))
# Adding explote and percentage values
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month, using the Object-Oriented Interface
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%', shadow=True, counterclock=False, startangle=90)
ax.axis('equal')

plt.show()
```
